```python
import pandas as pd
import seaborn as sns
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
data_path = '/content/drive/MyDrive/PDS_Assignment1/Q1/data_clean/cleaneddata.csv'
df = pd.read_csv(data_path)
```

```python
df.columns
```

```python
# Bar Plot of Frailty Distribution
plt.figure(figsize=(6, 4))
sns.countplot(x='Frailty', data=df,palette={'N': 'blue', 'Y': 'red'})
plt.title('Distribution of Frailty')
plt.xlabel('Frailty')
plt.ylabel('Count')

plt.savefig("/content/drive/MyDrive/PDS_Assignment1/Q1/results/distributionoffrality.png")
plt.show()
```

```python
# Box Plot of Age and GripStrength Grouped by Frailty
plt.subplot(1, 3, 3)
sns.boxplot(x='Frailty', y='Age', data=df, palette={'N': 'blue', 'Y': 'red'})
plt.title('Box Plot of Age Grouped by Frailty')
plt.xlabel('Frailty')
plt.ylabel('Age')

plt.tight_layout()

plt.savefig("/content/drive/MyDrive/PDS_Assignment1/Q1/results/boxplot of age.png")
plt.show()
```
