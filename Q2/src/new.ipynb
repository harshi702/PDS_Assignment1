```python
import pandas as pd
import seaborn as sns
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
data_path = '/content/drive/MyDrive/PDS_Assignment1/Q2/data_clean/StudentsPerformance.csv'
df = pd.read_csv(data_path)
```

```python
# Group by 'gender' and calculate the mean scores for each subject
gender_scores = df.groupby("gender")[["math score", "reading score", "writing score"]].mean()

# Create a bar chart
ax = gender_scores.plot(kind="bar", color=["darkblue", "skyblue", "purple"])

# Add labels and title
plt.xlabel("Gender")
plt.ylabel("Average Score")
plt.title("Student Performance by Gender")
plt.xticks(rotation=0)
plt.savefig("/content/drive/MyDrive/PDS_Assignment1/Q2/results/StudentsPerformance.png")
```

```python
values=df['parental level of education'].value_counts()
```

```python
plt.figure(figsize=(10,6))
explode=[0,0.3,0,0,0,0.4]
plt.pie(data=df,x=values,autopct="%.2f",normalize=True,shadow=True,explode=explode)

legend=plt.legend(df['parental level of education'],loc='upper right')
legend.set_bbox_to_anchor((1.3,1.3))


plt.savefig("/content/drive/MyDrive/PDS_Assignment1/Q2/results/levelofeducation.png")
plt.show()
```

```python
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True)
plt.savefig("/content/drive/MyDrive/PDS_Assignment1/Q2/results/haetmap.png")
```

```python
for item in df.columns[-3:]:
    sns.boxplot(
        x="test preparation course",
        y=item,
        hue="gender",
        palette="Set2",  # Use a color palette with distinct colors for male and female
        data=df
    )
    plt.title(f"{item} vs test preparation course", loc="left")

    plt.savefig("/content/drive/MyDrive/PDS_Assignment1/Q2/results/boxplotofscores.png")
    plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(20,6))
plt.subplot(1, 4, 1)
plt.title('MATH SCORES')
sns.violinplot(y='math score',data=df,color='darkblue',linewidth=3)
plt.subplot(1, 4, 2)
plt.title('READING SCORES')
sns.violinplot(y='reading score',data=df,color='skyblue',linewidth=3)
plt.subplot(1, 4, 3)
plt.title('WRITING SCORES')
sns.violinplot(y='writing score',data=df,color='purple',linewidth=3)
plt.savefig("/content/drive/MyDrive/PDS_Assignment1/Q2/results/violinplotscores.png")
plt.show()
```
